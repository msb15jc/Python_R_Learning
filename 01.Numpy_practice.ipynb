# Speed Test

```python
import numpy as np
from datetime import datetime

a = np.random.randn(100)
b = np.random.randn(100)
T = 100000

def slow_dot_product(a,b):
    result = 0
    for e,f in zip(a,b):
        result += e * f
    return result

t0 = datetime.now()

for t in range(T):
    slow_dot_product(a,b)
dt1 = datetime.now() - t0

t0 = datetime.now()
for t in range(T):
    a.dot(b)
dt2 = datetime.now() - t0

print("dt1 / dt2: ", dt1.total_seconds() / dt2.total_seconds())
```

## calculate PDF, CDF under the normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
x = np.linspace(-6,6, 1000)
```

```python
# for std, loc = mean, scale = std
fx = norm.pdf(x, loc = 0, scale = 1)
```

```python
plt.plot(x, fx);
```

```python
Fx = norm.cdf(x, loc = 0, scale = 1)
```

```python
plt.plot(x, Fx);
```

```python
# log pdf
logfx = norm.logpdf(x, loc = 0, scale = 1)
```

```python
plt.plot(x, logfx);
```

## scipy for convolution

```python
from PIL import Image
```

```python
im = Image.open('037.jpg')
```

```python
im
```

```python
# convert graphs into grey scales

gray = np.mean(im, axis = 2)
```

```python
x = np.linspace(-6,6, 50)
```

```python
fx = norm.pdf(x, loc = 0, scale = 1)
```

```python
filt = np.outer(fx, fx)
```

```python
plt.imshow(filt, cmap = 'gray')
```

```python
from scipy.signal import convolve2d
```

```python
out = convolve2d(gray, filt)
```

```python
plt.subplot(1,2,1)
plt.imshow(gray, cmap = "gray")
plt.subplot(1,2,2)
plt.imshow(out, cmap = 'gray')
```
